# Supervised Learning - a simple, linear case 

## Setup

```
x --->>>---[LINEAR]--->>>--- y

x = [1, 2, 3, 4]
y = [2, 4, 6, ?]
y_hat = x * w + b

Here, 
x => input vector
y => target/label vector
y_hat => predicted vector
w, b => model parameter to learn, w is weights vector and b is bias

assuming b = 0, the loss function:

loss (after seeing one training data) = (y_hat - y)^2
loss (total, i.e., after all training data) = (1/N) * sum[(y_hat - y)^2]
```

## Implementation

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]
````

````python
# randonmly initialise weight w
w = 1.0
````

````python
# define forward-pass funtion
def forward(x):
    return round((x * w), 2)
````

````python
# define loss function
def loss(x, y):
    y_hat = forward(x)
    return round(((y_hat - y) * (y_hat - y)), 2)
````

````python
# training 
w_list = []
mse_list = []
for w in np.arange(0.0, 4.1, 0.1):
    print("\n", "W =", round(w, 2))
    print("\t", "x_val", "\t", "y_val", "\t", "y_hat", "\t", "l_current")
    l_sum = 0
    for x_val, y_val in zip(x_data, y_data):
        y_hat = forward(x_val)
        l_current = loss(x_val, y_val)
        l_sum += l_current
        print("\t", x_val, "\t", y_val, "\t", y_hat, "\t", l_current)
    mse_error = round((l_sum / 3), 2)
    print("\n", "MSE =", mse_error)
    w_list.append(w)
    mse_list.append(mse_error)
````

````python
# plot loss vs. weight
plt.plot(w_list, mse_list)
plt.ylabel('Loss')
plt.xlabel('Weight')
plt.show()
````
